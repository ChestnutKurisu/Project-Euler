```python
def number_splittings(s: str, max_split_term_length: int, max_term: int):
    match len(s):
        case 0: return []
        case 1: return [[s]]
        case 2: return [[s], [s[0], s[1]]]
    
    total_possibilities = []
    for i in range(max_split_term_length):
        if int(s[:i + 1]) > max_term:
            break
        total_possibilities.extend([[s[:i + 1], *sub_possibility] for sub_possibility in number_splittings(s[i + 1:], min(len(s) - i - 1, max_split_term_length), max_term)])
        if len(s[i + 1:]) >= 3 and int(s[i + 1:]) < max_term:
            total_possibilities.append([s[:i + 1], s[i + 1:]])
    
    return total_possibilities
```

```python
def T(N):
    current_sum = 0
    for i in range(1, int(N ** 0.5) + 1):
        n = i ** 2
        for possibility in number_splittings(str(n), len(str(i)), i):
            if len(possibility) <= 1:
                continue
            if i == sum(int(m) for m in possibility):
                print('S-number:', n, possibility)
                current_sum += n
                break
                
    return current_sum
```
